```python
import time
import requests
import pprint as pp


data_list=[]
headers = {'Accept': 'application/json'}
for i in range(0,120):
    r=requests.get('http://api.open-notify.org/iss-now.json',headers=headers)
    time.sleep(10)
    data=r.json()
    pp.pprint(data)
    data_list.append(data)
```

```python
import matplotlib.pyplot as plt
latitudes=[]
longitudes=[]
timestamps=[]
for i in range(len(data_list)):
    lat = float(data_list[i]['iss_position']['latitude'])
    long = float(data_list[i]['iss_position']['longitude'])  # Fixed typo here
    timestamp = int(data_list[i]['timestamp'])
    latitudes.append(lat)
    longitudes.append(long)
    timestamps.append(timestamp)
    
plt.figure(figsize=(10, 6))
plt.plot(timestamps, longitudes)
plt.xlabel('Timestamp')
plt.ylabel('Longitude')
plt.title('ISS Longitude Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(timestamps, latitudes)
plt.xlabel('Timestamp')
plt.ylabel('Latitude')
plt.title('ISS Latitude Over Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(latitudes,longitudes)
plt.xlabel('Latitudes')
plt.ylabel('Longitudes')
plt.title('Coordinates Visualization')
plt.show()
```
